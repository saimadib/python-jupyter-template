### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

def plot_scatter_with_annotations(x, y, title="Scatter Plot with Annotations"):
    """
    Creates a scatter plot and annotates each point with its (x, y) coordinates.

    Parameters:
    - x (list of numeric): x-coordinates of the points.
    - y (list of numeric): y-coordinates of the points.
    - title (str): Title of the scatter plot.

    Returns:
    - None: Displays the scatter plot with annotations.

    Example:
        >>> plot_scatter_with_annotations([1,2,3], [4,5,6])
    """
    logger.debug("Entering plot_scatter_with_annotations function.")
    fig, ax = plt.subplots(figsize=(8, 6))
    logger.debug(f"Plotting scatter points x={x}, y={y}.")
    ax.scatter(x, y, s=100, alpha=0.75)
    for xi, yi in zip(x, y):
        label = f"({xi}, {yi})"
        logger.debug(f"Annotating point {label}.")
        ax.annotate(label, (xi, yi), textcoords="offset points", xytext=(5,5), ha='center')
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True, linestyle='-.', linewidth=0.5)
    plt.tight_layout()
    logger.info("Displaying scatter plot with annotations.")
    plt.show()
    logger.debug("Exiting plot_scatter_with_annotations function.")
    
def plot_scatter_with_annotations(x, y, title="Scatter Plot with Annotations"):
    """
    Creates a scatter plot and annotates each point with its (x, y) coordinates.

    Parameters:
    - x (list of numeric): x-coordinates of the points.
    - y (list of numeric): y-coordinates of the points.
    - title (str): Title of the scatter plot.

    Returns:
    - None: Displays the scatter plot with annotations.

    Example:
        >>> plot_scatter_with_annotations([1,2,3], [4,5,6])
    """
    logger.debug("Entering plot_scatter_with_annotations function.")
    fig, ax = plt.subplots(figsize=(8, 6))
    logger.debug(f"Plotting scatter points x={x}, y={y}.")
    ax.scatter(x, y, s=100, alpha=0.75)
    for xi, yi in zip(x, y):
        label = f"({xi}, {yi})"
        logger.debug(f"Annotating point {label}.")
        ax.annotate(label, (xi, yi), textcoords="offset points", xytext=(5,5), ha='center')
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True, linestyle='-.', linewidth=0.5)
    plt.tight_layout()
    logger.info("Displaying scatter plot with annotations.")
    plt.show()
    logger.debug("Exiting plot_scatter_with_annotations function.")
```
